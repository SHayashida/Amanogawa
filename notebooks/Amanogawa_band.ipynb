```python
# Create star distribution maps for the provided images.
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import blob_log
import pandas as pd
from pathlib import Path


image_paths = [
    "/content/IMG_5990.jpg",
    "/content/IMG_5991.jpg",
]

all_summaries = []

for image_path in image_paths:
    p = Path(image_path)
    # Load and convert to grayscale array
    img = Image.open(image_path).convert("L")
    data = np.array(img)

    # Detect bright spots (stars)
    # Parameters tuned gently for DSLR night sky shots
    blobs = blob_log(data, max_sigma=6, num_sigma=12, threshold=0.05)
    # blobs: rows of (y, x, radius)
    xs = blobs[:,1] if len(blobs)>0 else np.array([])
    ys = blobs[:,0] if len(blobs)>0 else np.array([])
    rs = blobs[:,2] if len(blobs)>0 else np.array([])

    # Save coordinates
    df = pd.DataFrame({"x": xs, "y": ys, "radius": rs})
    csv_path = f"/content/{p.stem}_star_coords.csv"
    df.to_csv(csv_path, index=False)

    # Basic summary stats
    summary = {
        "image": p.name,
        "width_px": data.shape[1],
        "height_px": data.shape[0],
        "num_stars_detected": len(df),
        "mean_radius_px": float(df["radius"].mean()) if len(df)>0 else 0.0,
    }
    all_summaries.append(summary)

    # 1) Scatter map of detected stars
    plt.figure(figsize=(8,8))
    plt.imshow(data, cmap="gray", origin="lower")
    for x, y, r in zip(xs, ys, rs):
        plt.gca().add_patch(plt.Circle((x, y), r*1.5, fill=False))
    plt.title(f"Detected Stars (scatter) — {p.name}")
    plt.xlabel("x (pixels)")
    plt.ylabel("y (pixels)")
    plt.tight_layout()
    plt.show()

    # 2) Density map via 2D histogram
    if len(df) > 0:
        # make a 2D histogram heatmap
        bins_x = max(10, data.shape[1]//100)  # roughly 100 px per bin
        bins_y = max(10, data.shape[0]//100)
        plt.figure(figsize=(8,8))
        plt.hist2d(xs, ys, bins=[bins_x, bins_y])
        plt.title(f"Star Density (2D histogram) — {p.name}")
        plt.xlabel("x (pixels)")
        plt.ylabel("y (pixels)")
        plt.colorbar(label="Count per bin")
        plt.tight_layout()
        plt.show()

# Show summary table
summary_df = pd.DataFrame(all_summaries)
display(summary_df)

# Provide the list of CSV outputs so the user can download them
[f"/content/{Path(p).stem}_star_coords.csv" for p in image_paths]
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from pathlib import Path
from skimage.feature import blob_log
from scipy.spatial import cKDTree
from scipy.optimize import curve_fit
import json, math, os, random

np.random.seed(42); random.seed(42)

IMAGE_PATH = "/content/drive/MyDrive/Colab Notebooks/天の川/IMG_5991.jpg"
COORDS_CSV = "/content/drive/MyDrive/Colab Notebooks/天の川/IMG_5991_star_coords.csv"  # 無ければ自動検出します

# --- I/O & 基本 ---
def load_image_gray(path):
    img = Image.open(path).convert("L")
    data = np.array(img)
    H, W = data.shape
    return data, (W, H)

def detect_stars_log(data, max_sigma=6, num_sigma=12, threshold=0.05):
    blobs = blob_log(data, max_sigma=max_sigma, num_sigma=num_sigma, threshold=threshold)
    if len(blobs) == 0:
        return pd.DataFrame(columns=["x","y","r"])
    return pd.DataFrame({"x": blobs[:,1], "y": blobs[:,0], "r": blobs[:,2]})

def load_or_detect(coords_csv, image_path, detect_threshold=0.05):
    data, (W,H) = load_image_gray(image_path)
    p = Path(coords_csv)
    if p.exists():
        df = pd.read_csv(p)
        print(f"[INFO] loaded coords from {p}, N={len(df)}")
    else:
        df = detect_stars_log(data, threshold=detect_threshold)
        df.to_csv(coords_csv, index=False)
        print(f"[INFO] detected stars by LoG, N={len(df)}, saved to {coords_csv}")
    m = (df["x"].between(0, W-1)) & (df["y"].between(0, H-1))
    df = df[m].reset_index(drop=True)
    return data, (W,H), df

# --- 幾何・統計 ---
def nearest_neighbor_dist(points):
    tree = cKDTree(points)
    dists, _ = tree.query(points, k=2)
    return dists[:,1]

def two_point_correlation(points, W, H, r_bins, maxN=4000):
    """ξ(r) ≈ DD/RR - 1（Poisson参照）。DR省略の簡易版。"""
    P = points
    if len(P) > maxN:
        idx = np.random.choice(len(P), maxN, replace=False)
        P = P[idx]
    tree = cKDTree(P)
    DD = np.zeros(len(r_bins)-1)
    for i in range(len(P)):
        idxs = tree.query_ball_point(P[i], r_bins[-1])
        d = np.linalg.norm(P[idxs]-P[i], axis=1); d = d[d>0]
        h,_ = np.histogram(d, bins=r_bins); DD += h
    DD /= 2.0
    # Poisson参照
    R = np.column_stack([np.random.uniform(0, W, len(P)),
                         np.random.uniform(0, H, len(P))])
    treeR = cKDTree(R)
    RR = np.zeros(len(r_bins)-1)
    for i in range(len(R)):
        idxs = treeR.query_ball_point(R[i], r_bins[-1])
        d = np.linalg.norm(R[idxs]-R[i], axis=1); d = d[d>0]
        h,_ = np.histogram(d, bins=r_bins); RR += h
    RR /= 2.0
    xi = DD/(RR+1e-9) - 1.0
    r_centers = 0.5*(r_bins[1:]+r_bins[:-1])
    return r_centers, xi

def boxcount_fractal_dim(points, W, H, exp_from=-1.0, exp_to=-2.2, steps=10):
    X = np.column_stack([points[:,0]/W, points[:,1]/H])
    eps_list = np.logspace(exp_from, exp_to, steps)
    Ns = []
    for eps in eps_list:
        bins = int(np.ceil(1/eps))
        grid = np.floor(X*bins).astype(int)
        grid[grid==bins] = bins-1
        Ns.append(len(np.unique(grid, axis=0)))
    eps = np.array(eps_list); Ns = np.array(Ns)
    D = np.polyfit(np.log(1/eps), np.log(Ns), 1)[0]
    return D, eps, Ns

def pca_principal_axis(points, W, H, bins_x=60):
    bins_y = int(bins_x * H / W)
    hist, xedges, yedges = np.histogram2d(points[:,0], points[:,1],
                                          bins=[bins_x, bins_y],
                                          range=[[0,W],[0,H]])
    xc = 0.5*(xedges[1:]+xedges[:-1]); yc = 0.5*(yedges[1:]+yedges[:-1])
    XX, YY = np.meshgrid(xc, yc, indexing='xy'); weights = hist.T
    xw = np.average(XX, weights=weights); yw = np.average(YY, weights=weights)
    dx = XX-xw; dy = YY-yw
    C = np.array([[np.average(dx*dx, weights=weights), np.average(dx*dy, weights=weights)],
                  [np.average(dx*dy, weights=weights), np.average(dy*dy, weights=weights)]])
    vals, vecs = np.linalg.eig(C)
    v = vecs[:, np.argmax(vals)]
    ang = math.degrees(math.atan2(v[1], v[0]))
    return ang, (xw,yw)

def rotate_points(points, angle_deg, center):
    th = math.radians(angle_deg)
    R = np.array([[math.cos(th), -math.sin(th)],[math.sin(th), math.cos(th)]])
    return (points-center)@R.T + center

# --- フォトメトリ（インデックス返す版） ---
def simple_aperture_mags_with_indices(img, points, r=2, r_in=6, r_out=10, max_samples=15000):
    H, W = img.shape
    if len(points) > max_samples:
        idx_sample = np.random.choice(len(points), max_samples, replace=False)
        pts_sample = points[idx_sample]
    else:
        idx_sample = np.arange(len(points))
        pts_sample = points
    mags = []; valid_idx = []
    for i,(x,y) in enumerate(pts_sample):
        x=int(round(x)); y=int(round(y))
        x0=max(0,x-r); x1=min(W,x+r+1); y0=max(0,y-r); y1=min(H,y+r+1)
        stamp = img[y0:y1, x0:x1]; signal = np.sum(stamp)
        yin0=max(0,y-r_out); yin1=min(H,y+r_out+1)
        xin0=max(0,x-r_out); xin1=min(W,x+r_out+1)
        ann = img[yin0:yin1, xin0:xin1].astype(float)
        yy,xx = np.ogrid[yin0:yin1, xin0:xin1]
        ann = ann[~(((xx-x)**2+(yy-y)**2) <= r_in**2)]
        bkg = np.median(ann) if ann.size>0 else 0.0
        flux = signal - bkg*stamp.size
        if flux>0:
            mags.append(-2.5*np.log10(flux)); valid_idx.append(idx_sample[i])
    return np.array(mags), np.array(valid_idx)
```

```python
thresholds = np.linspace(0.03, 0.08, 10)
r_bins = np.linspace(5, min(W,H)/3, 22)
rows = []
for th in thresholds:
    df_th = detect_stars_log(data, threshold=th)
    P = df_th[["x","y"]].to_numpy()
    if len(P) < 200:
        rows.append({"threshold": th, "N": len(P), "D": np.nan, "xi_mean": np.nan})
        continue
    D,_,_ = boxcount_fractal_dim(P, W, H)
    rc, xi = two_point_correlation(P, W, H, r_bins, maxN=3500)
    rows.append({"threshold": th, "N": len(P), "D": float(D), "xi_mean": float(np.nanmean(xi))})

sweep = pd.DataFrame(rows); display(sweep)

fig, ax = plt.subplots(1,3, figsize=(15,4))
ax[0].plot(sweep["threshold"], sweep["N"], marker="o"); ax[0].set_title("N vs threshold")
ax[1].plot(sweep["threshold"], sweep["D"], marker="o"); ax[1].set_title("Fractal D vs threshold")
ax[2].plot(sweep["threshold"], sweep["xi_mean"], marker="o"); ax[2].set_title("Mean ξ vs threshold")
for a in ax: a.set_xlabel("threshold")
plt.tight_layout(); plt.show()

sweep.to_csv("threshold_sweep_summary.csv", index=False)
print("saved: threshold_sweep_summary.csv")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import math, json

# ====== 必要なら定義：主軸推定と回転 ======
def pca_principal_axis(points, W, H, bins_x=60):
    bins_y = int(bins_x * H / W)
    hist, xedges, yedges = np.histogram2d(points[:,0], points[:,1],
                                          bins=[bins_x, bins_y],
                                          range=[[0,W],[0,H]])
    xc = 0.5*(xedges[1:]+xedges[:-1]); yc = 0.5*(yedges[1:]+yedges[:-1])
    XX, YY = np.meshgrid(xc, yc, indexing='xy'); weights = hist.T
    xw = np.average(XX, weights=weights); yw = np.average(YY, weights=weights)
    dx = XX-xw; dy = YY-yw
    C = np.array([[np.average(dx*dx, weights=weights), np.average(dx*dy, weights=weights)],
                  [np.average(dx*dy, weights=weights), np.average(dy*dy, weights=weights)]])
    vals, vecs = np.linalg.eig(C)
    v = vecs[:, np.argmax(vals)]
    ang = math.degrees(math.atan2(v[1], v[0]))
    return ang, (xw, yw)

def rotate_points(points, angle_deg, center):
    th = math.radians(angle_deg)
    R = np.array([[math.cos(th), -math.sin(th)],[math.sin(th), math.cos(th)]])
    return (points-center)@R.T + center

# ====== モデル関数（背景項つき） ======
def gauss_bg(x, A, mu, sigma, B):
    return A*np.exp(-(x-mu)**2/(2*sigma**2)) + B

def lorentz_bg(x, A, x0, gamma, B):
    return A*(gamma**2/((x-x0)**2 + gamma**2)) + B

# ====== 1) 主軸 → 直交プロファイル（ヒスト） ======
ang_deg, center = pca_principal_axis(points, W, H, bins_x=60)
P_rot = rotate_points(points, -ang_deg, np.array(center))
y_perp = P_rot[:,1]

bins = 140                         # 論文用にやや細かめ
hist, edges = np.histogram(y_perp, bins=bins)
centers = 0.5*(edges[1:]+edges[:-1])

# 背景推定：両端の下位分位を使用（裾の影響を避ける）
edge_frac = 0.15
k = int(max(3, bins*edge_frac))
bg_samples = np.concatenate([hist[:k], hist[-k:]])
B0 = np.median(bg_samples)         # 背景の初期推定

# ====== 2) フィット範囲の自動決定（山の付近に限定） ======
peak_idx = np.argmax(hist)
peak = hist[peak_idx]
amp0 = max(peak - B0, 1.0)
mu0  = centers[peak_idx]
sig0 = np.std(y_perp)

# 山の部分：背景+0.2*amp 以上の区間
mask_peak = hist >= (B0 + 0.2*amp0)
x_fit = centers[mask_peak]
y_fit = hist[mask_peak]

# 初期値
p0_g = (amp0, mu0, sig0, B0)
p0_l = (amp0, mu0, max(sig0,1.0), B0)

# フィット
pg, _ = curve_fit(gauss_bg, x_fit, y_fit, p0=p0_g, maxfev=20000)
pl, _ = curve_fit(lorentz_bg, x_fit, y_fit, p0=p0_l, maxfev=20000)

# FWHM（px）
FWHM_g = 2.3548*abs(pg[2])
FWHM_l = 2*abs(pl[2])

# ====== 3) 実データ半値幅（empirical FWHM） ======
# 背景を差し引いて正規化し、ピーク/2 を交差する左右点を線形補間で求める
profile_sub = hist - B0
profile_sub[profile_sub < 0] = 0
peak_val = profile_sub[peak_idx]
half = peak_val * 0.5

# 左右の交差点探索（線形補間）
def half_cross_x(centers, prof, start, direction):
    i = start
    while 0 <= i+direction < len(prof) and prof[i+direction] > half:
        i += direction
    j = i + direction
    if j < 0 or j >= len(prof):
        return centers[i]  # フォールバック
    # 線形補間
    x1, y1 = centers[i], prof[i]
    x2, y2 = centers[j], prof[j]
    if y2 == y1:
        return x1
    return x1 + (half - y1) * (x2 - x1) / (y2 - y1)

x_left  = half_cross_x(centers, profile_sub, peak_idx, -1)
x_right = half_cross_x(centers, profile_sub, peak_idx, +1)
FWHM_emp = x_right - x_left

# ====== 4) 図を重ね描き（論文体裁） ======
xplot = np.linspace(centers.min(), centers.max(), 1500)
plt.figure(figsize=(9,5.2))

# 観測
plt.plot(centers, hist, "k.", ms=4, label="Observed profile (bin counts)")

# フィット
plt.plot(xplot, gauss_bg(xplot, *pg), "-",  lw=2, label=f"Gaussian+bg (FWHM ≈ {FWHM_g:.1f} px)")
plt.plot(xplot, lorentz_bg(xplot, *pl), "--", lw=2, label=f"Lorentzian+bg (FWHM ≈ {FWHM_l:.1f} px)")

# 背景線
plt.axhline(B0, lw=1, linestyle=":", label=f"Estimated background (≈{B0:.1f})")

# 実データ半値幅のシェーディング
plt.fill_between([x_left, x_right],
                 [0, 0],
                 [max(hist)*1.05, max(hist)*1.05],
                 alpha=0.12, label=f"Empirical FWHM ≈ {FWHM_emp:.1f} px")

plt.xlabel("Perpendicular coordinate y′ (px)")
plt.ylabel("Star counts per bin")
plt.title("Milky Way band cross-section with background and FWHM")
plt.xlim(centers.min(), centers.max())
plt.ylim(0, max(hist)*1.05)
plt.legend()
plt.tight_layout()
plt.show()

# ====== 5) 角度スケールに換算（任意：plate scale がある場合） ======
PLATE_SCALE_ARCSEC = 66.959  # 既知の plate scale があれば設定。無ければ None
out = {
    "principal_axis_angle_deg_from_x": float(ang_deg),
    "fit_range_bins": int(mask_peak.sum()),
    "background_median": float(B0),
    "FWHM_gaussian_px": float(FWHM_g),
    "FWHM_lorentz_px": float(FWHM_l),
    "FWHM_empirical_px": float(FWHM_emp),
}
if PLATE_SCALE_ARCSEC is not None:
    px2deg = PLATE_SCALE_ARCSEC/3600.0
    out.update({
        "FWHM_gaussian_deg": float(FWHM_g*px2deg),
        "FWHM_lorentz_deg":  float(FWHM_l*px2deg),
        "FWHM_empirical_deg":float(FWHM_emp*px2deg),
    })

with open("band_width_fit_summary.json","w") as f:
    json.dump(out, f, indent=2)

print("Saved band_width_fit_summary.json")
for k,v in out.items():
    print(f"{k}: {v}")
```
